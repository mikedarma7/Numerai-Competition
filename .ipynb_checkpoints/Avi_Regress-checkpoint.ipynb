# Aviation Regression Problem

**In brief, we are going to tackle the regression problem "Predicting Total Injuries" using Stochastic Gradient Descent**

There are several steps that we will undergo: 
>1. One-Hot Encoding

```python
import pandas as pd
import seaborn as sb
import matplotlib.pyplot as plt
```

```python
data_clean = pd.read_csv("DataClass/data_clean.csv")
data_clean.head()
```

```python
data_clean.columns
```

## 1. One-Hot Encoding

We need to transform categorical variables into numerical variable ('1' or '0' matrix)

### 1. Encoding 10 Makes

```python
data_make = pd.DataFrame(data_clean)
data_make["Make"] = pd.Categorical(data_make["Make"])
make = pd.get_dummies(data_make["Make"])
data_make = pd.concat([data_make,make],axis = 1)
```

```python
data_make.head()
```

### 2. Encode Division

```python
data_div = pd.DataFrame(data_make)
data_div["Division"] = pd.Categorical(data_div["Division"])
division = pd.get_dummies(data_div["Division"])
data_div = pd.concat([data_div,division],axis = 1)
```

```python
data_div.head()
```

### 3. Encode Broad.Phase.of.Flight

```python
data_bop = pd.DataFrame(data_div)
data_bop["Broad.Phase.of.Flight"] = pd.Categorical(data_bop["Broad.Phase.of.Flight"])
phase = pd.get_dummies(data_bop["Broad.Phase.of.Flight"])
data_bop = pd.concat([data_bop,phase],axis = 1)
data_bop.head()
```

### 4. Encode Aircraft Damage

Group Minor with Substantial together

```python
# for x in range (len(data_bop["Aircraft.Damage"].unique())):
#     data_bop["Aircraft.Damage"][data_bop["Aircraft.Damage"] == "Minor"] = "Substantial"
```

```python
data_dam = pd.DataFrame(data_bop)
data_dam["Aircraft.Damage"] = pd.Categorical(data_dam["Aircraft.Damage"])
damage = pd.get_dummies(data_dam["Aircraft.Damage"])
data_dam = pd.concat([data_dam,damage],axis = 1)
data_dam.head()
```

### 5. Encode Weather Condition

```python
data_wc = pd.DataFrame(data_dam)
data_wc["Weather.Condition"] = pd.Categorical(data_wc["Weather.Condition"])
weather = pd.get_dummies(data_wc["Weather.Condition"])
data_comb = pd.concat([data_wc,weather],axis = 1)
data_comb.head()
```

## Dropping the unecessary column!

Drop Division, Aircraft.Damage, Make, Model, Amateur.Built, Weather.Condition, Broad.Phase.of.Flight!

Drop Total.Uninjured, Total.Injured, Total.Passenger, %Injury as well as it's not necessary for classification.

Note that Model is too many to be encoded and may cause overfitting in the end!

```python
data_reg = data_comb.drop(["Division","Aircraft.Damage","Make","Model","Amateur.Built","Weather.Condition","Broad.Phase.of.Flight",
                            "Total.Uninjured"],axis=1)
data_reg.head()
```

```python
data_reg.columns
```

Convert any floats into int64 dtypes.

```python
data_reg.dtypes
```

```python
data_reg["Number.of.Engines"] = data_reg["Number.of.Engines"].astype('int64')
data_reg["Total.Injured"] = data_reg["Total.Injured"].astype('int64')
data_reg["Total.Passenger"] = data_reg["Total.Passenger"].astype('int64')
data_reg.head()
```

```python
data_reg = data_reg.drop(["STANDING","Air Tractor","UNK","Minor"],axis=1)
data_reg.head()
```

## 2. Train-Test Split

```python
from sklearn.model_selection import train_test_split
```

Splitting the Train and Test data, test size = 25%

```python
reg_x = data_reg.drop(["Total.Injured"],axis = 1) #anything except Severity
reg_y = data_reg["Total.Injured"]

reg_x_train, reg_x_test, reg_y_train, reg_y_test = train_test_split(reg_x, reg_y, test_size=0.25,random_state=21)
print(reg_x_train.__len__(), reg_y_train.__len__())
```

## 3. Fitting Linear Regression Model

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
```

```python
linreg = LinearRegression()       
linreg.fit(reg_x_train, reg_y_train)        

print('Intercept of Regression \t: b = ', linreg.intercept_)
print('Coefficients of Regression \t: a = ', linreg.coef_)
print()

print(pd.DataFrame(list(zip(reg_x_train.columns, linreg.coef_[0])), columns = ["Predictors", "Coefficients"]))
print()
```

### Predicting Response corresponding to Predictors

```python
reg_y_train_pred = linreg.predict(reg_x_train)
reg_y_test_pred = linreg.predict(reg_x_test)
```

```python
f, axes = plt.subplots(1, 2, figsize=(12, 6))
axes[0].scatter(reg_y_train, reg_y_train_pred, color = "blue")
axes[0].plot(reg_y_train, reg_y_train, 'w-', linewidth = 1)
axes[0].set_xlabel("True values of the Response Variable (Train)")
axes[0].set_ylabel("Predicted values of the Response Variable (Train)")
axes[1].scatter(reg_y_test, reg_y_test_pred, color = "green")
axes[1].plot(reg_y_test, reg_y_test, 'w-', linewidth = 1)
axes[1].set_xlabel("True values of the Response Variable (Test)")
axes[1].set_ylabel("Predicted values of the Response Variable (Test)")
plt.show()
```

```python
# Check the Goodness of Fit (on Train Data)
print("Goodness of Fit of Model \tTrain Dataset")
print("Explained Variance (R^2) \t:", linreg.score(reg_x_train, reg_y_train))
print("Mean Squared Error (MSE) \t:", mean_squared_error(reg_y_train, reg_y_train_pred))
print()

# Check the Goodness of Fit (on Test Data)
print("Goodness of Fit of Model \tTest Dataset")
print("Explained Variance (R^2) \t:", linreg.score(reg_x_test, reg_y_test))
print("Mean Squared Error (MSE) \t:", mean_squared_error(reg_y_test, reg_y_test_pred))
print()
```

## 2. Stochastic Gradient

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(reg_x_train)  # Don't cheat - fit only on training data
X_train = scaler.transform(reg_x_train)
X_test = scaler.transform(reg_x_test)  # apply same transformation to test data
```

```python
from sklearn.linear_model import SGDClassifier
clf = SGDClassifier(loss="hinge", penalty="l2", max_iter=20)
clf.fit(X_train, reg_y_train)
```

```python
reg_y_train_pred1 = clf.predict(X_train)
reg_y_test_pred1 = clf.predict(X_test)
```

```python
# Check the Goodness of Fit (on Train Data)
print("Goodness of Fit of Model \tTrain Dataset")
print("Explained Variance (R^2) \t:", clf.score(X_train, reg_y_train))
print("Mean Squared Error (MSE) \t:", mean_squared_error(reg_y_train, reg_y_train_pred1))
print()

# Check the Goodness of Fit (on Test Data)
print("Goodness of Fit of Model \tTest Dataset")
print("Explained Variance (R^2) \t:", clf.score(X_test, reg_y_test))
print("Mean Squared Error (MSE) \t:", mean_squared_error(reg_y_test, reg_y_test_pred1))
print()
```

```python
l1_ratio = 0.5  # L1 weight in the Elastic-Net regularization

fig, axes = plt.subplots(3, 3)

# Set regularization parameter
for i, (C, axes_row) in enumerate(zip((1, 0.1, 0.01), axes)):
    # turn down tolerance for short training time
    clf_l1_LR = LogisticRegression(C=C, penalty='l1', tol=0.01, solver='saga')
    clf_l2_LR = LogisticRegression(C=C, penalty='l2', tol=0.01, solver='saga')
    clf_en_LR = LogisticRegression(C=C, penalty='elasticnet', solver='saga',
                                   l1_ratio=l1_ratio, tol=0.01)
    clf_l1_LR.fit(reg_x_train, reg_y_train)
    clf_l2_LR.fit(reg_x_train, reg_y_train)
    clf_en_LR.fit(reg_x_train, reg_y_train)

    coef_l1_LR = clf_l1_LR.coef_.ravel()
    coef_l2_LR = clf_l2_LR.coef_.ravel()
    coef_en_LR = clf_en_LR.coef_.ravel()

    # coef_l1_LR contains zeros due to the
    # L1 sparsity inducing norm

    sparsity_l1_LR = np.mean(coef_l1_LR == 0) * 100
    sparsity_l2_LR = np.mean(coef_l2_LR == 0) * 100
    sparsity_en_LR = np.mean(coef_en_LR == 0) * 100

    print("Train Data")
    print("C=%.2f" % C)
    print("{:<40} {:.2f}%".format("Sparsity with L1 penalty:", sparsity_l1_LR))
    print("{:<40} {:.2f}%".format("Sparsity with Elastic-Net penalty:",
                                  sparsity_en_LR))
    print("{:<40} {:.2f}%".format("Sparsity with L2 penalty:", sparsity_l2_LR))
    print("{:<40} {:.2f}".format("Score with L1 penalty:",
                                 clf_l1_LR.score(reg_x_train, reg_y_train)))
    print("{:<40} {:.2f}".format("Score with Elastic-Net penalty:",
                                 clf_en_LR.score(reg_x_train, reg_y_train)))
    print("{:<40} {:.2f}".format("Score with L2 penalty:",
                                 clf_l2_LR.score(reg_x_train, reg_y_train)))
    
    print("Test Data")
    print("C=%.2f" % C)
    print("{:<40} {:.2f}%".format("Sparsity with L1 penalty:", sparsity_l1_LR))
    print("{:<40} {:.2f}%".format("Sparsity with Elastic-Net penalty:",
                                  sparsity_en_LR))
    print("{:<40} {:.2f}%".format("Sparsity with L2 penalty:", sparsity_l2_LR))
    print("{:<40} {:.2f}".format("Score with L1 penalty:",
                                 clf_l1_LR.score(reg_x_test, reg_y_test)))
    print("{:<40} {:.2f}".format("Score with Elastic-Net penalty:",
                                 clf_en_LR.score(reg_x_test, reg_y_test)))
    print("{:<40} {:.2f}".format("Score with L2 penalty:",
                                 clf_l2_LR.score(reg_x_test, reg_y_test)))
    
```
