# Aviation Classification Problem

**In brief, we are going to tackle the classification problem "Severity of Aircraft Accident" using four models: Decision Tree, Random Forest, Logistic Regression, and Neural Network**

Let's say we are riding an aircraft and we don't have that much time to count how many passengers there are, but we do know the rest of the variables. Will the accident be considered as "Negligible" (Positive) or "Catastrophic" (Negative) risk?

There are several steps that we will undergo: 
>1. Handcrafting Features 
>2. One-Hot Encoding 
>3. Remove Dummy Variable Trap 
>4. Train-Test Split
>5. Fitting & Scoring Models:  Decision Tree, Random Forest, Log-reg, Neural Network

```python
import pandas as pd
import seaborn as sb
```

```python
data_clean = pd.read_csv("DataClass/data_clean.csv")
data_clean.head()
```

```python
data_clean.columns
```

## 1. Handcrafting Features

**We need to add a new categorical variable "Injury.Severity" based on Total Injuries!**

> Defining Injury Severity based on %Injury = Total Injury/Total Passenger

0-50% == Negligible (0)

51-100% == Catastrophic (1)

```python
insev = []
for i in range(len(data_clean)):
    insev.append(data_clean["Total.Injured"].iloc[i]/data_clean["Total.Passenger"].iloc[i]*100)
    
data_clean["%Injury"] = insev
data_clean.head()
```

```python
severity = []
for i in range(len(data_clean)):
    if data_clean["%Injury"].iloc[i] == 0 or data_clean["%Injury"].iloc[i] < 50:
        severity.append("Negligible")
    else:
        severity.append("Catastrophic")

data_clean["Severity"] = severity
data_clean.head()
```

```python
print(data_clean["Severity"].value_counts())
print("Total data:", data_clean["Severity"].count() + 1)
sb.catplot(y = "Severity", data = data_clean, kind = "count")
```

#### Switch Response "Negligible" to "1" and "Catastrophic" to "0 for later use (i.e. Neural Network TensorFlow model)

```python
data_clean = data_clean.replace(to_replace = {"Severity":{"Negligible":0,"Catastrophic":1}})
```

```python
data_clean.head()
```

## 2. One-Hot Encoding

We need to transform categorical variables into numerical variable ('1' or '0' matrix)

### 1. Encoding 10 Makes

```python
data_make = pd.DataFrame(data_clean)
data_make["Make"] = pd.Categorical(data_make["Make"])
make = pd.get_dummies(data_make["Make"])
data_make = pd.concat([data_make,make],axis = 1)
```

```python
data_make.head()
```

### 2. Encode Division

```python
data_div = pd.DataFrame(data_make)
data_div["Division"] = pd.Categorical(data_div["Division"])
division = pd.get_dummies(data_div["Division"])
data_div = pd.concat([data_div,division],axis = 1)
```

```python
data_div.head()
```

### 3. Encode Broad.Phase.of.Flight

```python
data_bop = pd.DataFrame(data_div)
data_bop["Broad.Phase.of.Flight"] = pd.Categorical(data_bop["Broad.Phase.of.Flight"])
phase = pd.get_dummies(data_bop["Broad.Phase.of.Flight"])
data_bop = pd.concat([data_bop,phase],axis = 1)
data_bop.head()
```

### 4. Encode Aircraft Damage

Group Minor with Substantial together

```python
data_dam = pd.DataFrame(data_bop)
data_dam["Aircraft.Damage"] = pd.Categorical(data_dam["Aircraft.Damage"])
damage = pd.get_dummies(data_dam["Aircraft.Damage"])
data_dam = pd.concat([data_dam,damage],axis = 1)
data_dam.head()
```

### 5. Encode Weather Condition

```python
data_wc = pd.DataFrame(data_dam)
data_wc["Weather.Condition"] = pd.Categorical(data_wc["Weather.Condition"])
weather = pd.get_dummies(data_wc["Weather.Condition"])
data_comb = pd.concat([data_wc,weather],axis = 1)
data_comb.head()
```

## Dropping the unecessary column!

Drop Division, Aircraft.Damage, Make, Model, Amateur.Built, Weather.Condition, Broad.Phase.of.Flight!

Drop Total.Uninjured, Total.Injured, Total.Passenger, %Injury as well as it's not necessary for classification.

Note that Model is too many to be encoded and may cause overfitting in the end!

```python
data_class = data_comb.drop(["Division","Aircraft.Damage","Make","Model","Amateur.Built","Weather.Condition","Broad.Phase.of.Flight",
                            "Total.Uninjured","Total.Injured","Total.Passenger","%Injury"],axis=1)
data_class.head()
```

```python
data_class.columns
```

## 3. Avoid Dummy Variable Trap

**To avoid dummy variable trap, we must drop one encoded variable, let's say SATL, Cessna, Landing, Substantial**

Reason: When you can derive one variable from other variables, they are known to be multi-colinear. Here if you know values of california and georgia then you can easily infer value of new jersey state, i.e. california=0 and georgia=0. There for these state variables are called to be multi-colinear. In this situation linear regression won't work as expected. Hence you need to drop one column.

```python
data_class = data_class.drop(["NWED","Mooney","STANDING","Minor","UNK"],axis=1)
data_class.head()
```

```python
data_class.columns
```

Convert any floats into int64 dtypes.

```python
data_class.dtypes
```

```python
data_class["Number.of.Engines"] = data_class["Number.of.Engines"].astype('int64')
data_class.head()
```

## 4. Train-Test Split

```python
from sklearn.model_selection import train_test_split
```

Splitting the Train and Test data, test size = 25%

```python
class_x = data_class.drop(["Severity"],axis = 1) #anything except Severity
class_y = data_class["Severity"]

class_x_train, class_x_test, class_y_train, class_y_test = train_test_split(class_x, class_y, test_size=0.25,random_state=21)
print(class_x_train.__len__(), class_y_train.__len__())
```

Export the test and train data for x and y into csv for later use in separate Jupyter Notebook for prediction

```python
class_x_train.to_csv("DataClass/xtraindata.csv", index=False)
class_x_test.to_csv("DataClass/xtestdata.csv", index=False)
class_y_train.to_csv("DataClass/ytraindata.csv", index=False)
class_y_test.to_csv("DataClass/ytestdata.csv", index=False)
```

```python
class_x_train.dtypes
```

## 5. Model Fitting and Scoring

### a. Decision Tree Classification

```python
from sklearn.tree import DecisionTreeClassifier
import os
os.environ["PATH"] += os.pathsep + 'C:/Program Files (x86)/Graphviz2.38/bin/'
import graphviz
from sklearn.tree import export_graphviz
from sklearn.metrics import confusion_matrix
from os import system
```

#### Fitting the model and evaluating the model on each region

```python
dectree = DecisionTreeClassifier(max_depth=6)
model = dectree.fit(class_x_train, class_y_train)
print("Train data model score is: {}\n".format(dectree.score(class_x_train, class_y_train)))
print("Test data model score is: {}\n".format(dectree.score(class_x_test, class_y_test)))
```

```python
treedot = export_graphviz(model,                                      # the model
                          feature_names = class_x_train.columns,          # the features 
                          out_file = None,                              # output file
                          filled = True,                                # node colors
                          rounded = True,                               # make pretty
                          special_characters = True, class_names = ["0","1"]) 
graphviz.Source(treedot)
```

Saving the Model to be used for prediction in a separate .ipynb

```python
import pickle
pickle.dump(dectree, open("DECModels/decmodelcombined.sav",'wb'))
```

#### Scoring Model using Confusion Matrix

```python
class_y_test_pred = dectree.predict(class_x_test)

sb.heatmap(confusion_matrix(class_y_test, class_y_test_pred), 
           annot = True, fmt=".0f", annot_kws={"size": 18},cmap = "PuRd")
```

**Computing F1 Score. F1 = 2(Precision x Recall)/(Precision + Recall)**

with Precision = TP/(TP+FP) and Recall = TP/(TP+FN)

Precision is how precise the predicted positive with the actual.
Recall is how good the model captures actual positives. 

Note: Accuracy can be largely contributed by a large number of True Negatives which in most business circumstances, we do not focus on much whereas False Negative and False Positive usually has business costs (tangible & intangible) thus F1 Score might be a better measure to use if we need to seek a balance between Precision and Recall AND there is an uneven class distribution (large number of Actual Negatives).

```python
from sklearn.metrics import f1_score
f1_score(class_y_test,class_y_test_pred,average='weighted')
```

Max F1 score is 1, but ~0.756 is good enough.

### 4. Neural Network using TensorFlow

```python
import tensorflow as tf
from tensorflow import keras
```

```python
dataset = tf.data.Dataset.from_tensor_slices((class_x_train.values,class_y_train.values))
```

```python
train_dataset = dataset.batch(256)
```

Defining the model with 2 hidden layers with 'relu' activation function.

```python
model = keras.Sequential([
    keras.layers.Dense(100,activation = 'relu'),
    keras.layers.Dense(50,activation = 'relu'),
    keras.layers.Dense(1,activation = 'sigmoid')
])

model.compile(
    optimizer = 'adam',
    loss = "binary_crossentropy",
    metrics = ['acc']
)
```

```python
model.fit(train_dataset,epochs = 2048,verbose = 2)
```

```python
model.evaluate(class_x_train.values,class_y_train.values)
```

```python
model.evaluate(class_x_test.values,class_y_test.values)
```

```python
model.save("KERASModels/kerasmodelcombined.h5")
```
